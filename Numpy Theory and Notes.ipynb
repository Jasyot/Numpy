# Mathamatical operations on array (elementwise):

The size of array should be the same in order to execute this. That is number of rows and number of colls should be same in all the array!

```python
# Addition/substraction:
a=np.random.rand(2,3)
b=np.random.rand(2,3)

c=a+b
c
```

```python
# Multiplication/division
a=np.random.randint(2,100,size=(2,3))
b=np.random.randint(20,1000,size=(2,3))
c=a*b
c
```

```python
# Element wise addition
a=np.random.rand(2,3)
print(a)
b=a+30000000000000000
print(b)
```

```python
# Multiplication of two array (not matrix multiplication, element wise)
a=np.ones((2,3))
print(a)
b=np.zeros((2,3))
print(b)
c=a*b
print(c)
```

```python
# Elementwise multiplication
a=np.random.rand(2,3)
print(a)
b=a*30000000000000000
print(b)
```

```python
# Squareroot:
arr=np.random.randint(10,20,size=(2,3))
arr=np.sqrt(arr)
print(arr)
```

```python
# exponent
np.exp(arr)
```

```python
# log
np.log(arr)
```

# Matrix multiplication of array

```python
a=np.random.rand(4,6)
b=np.random.rand(6,10)

mat_multi_ab=np.matmul(a,b)
print(mat_multi_ab)
```

# Indexing and Slicing in array

```python
# Defining array:

arr=np.array([[1,2,3,4],[2,3,4,5],[4,5,4,3]])
print("the shape is",arr.shape)
print("*"*25)
print("the dim is",arr.ndim)
print("*"*25)
print("the size is",arr.size)
print("*"*25)
print("the array is\n",arr)
```

```python
# INDEXING:
```

```python
arr[0]
```

```python
arr[0][0]
```

```python
# Adding np.nan to an array
arr=np.array([[1,2,3,4],[2,3,4,5],[4,5,4,3]],dtype='float') #if dtype is int, np.nan will not convert int to nan!

arr[0][0]=np.nan
print(arr)
```

```python
# SLICING:
```

```python
arr[1:,0:2]
```

```python
arr[0:2,1:3]
```
